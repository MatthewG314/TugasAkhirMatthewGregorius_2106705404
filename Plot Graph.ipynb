## SSIM

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = "C:\Matthew\Kuliah\Semester 8\Pengolahan Data\Progress code terbaru (4 apr)\Analisis2.xlsx" 
xls = pd.ExcelFile(file_path)

selected_range = pd.read_excel(xls, sheet_name=0, usecols="A:I", skiprows=0, nrows=24)

selected_range_clean = selected_range.apply(pd.to_numeric, errors='coerce').dropna()

x_values = selected_range_clean.iloc[:, 0]

y_columns = [7, 3, 5, 1]
std_columns = [8, 4, 6, 2]

colors = ['black', 'green', 'red', 'blue']

plt.figure(figsize=(10, 10))

for (y_col, std_col, color) in zip(y_columns, std_columns, colors):
    plt.errorbar(
        x_values, 
        selected_range_clean.iloc[:, y_col], 
        yerr=selected_range_clean.iloc[:, std_col], 
        fmt='o-', capsize=5, label=selected_range_clean.columns[y_col],
        color=color
    )


plt.ylabel("Values with Error Bars")
plt.legend(["Original Image", "Denoised Image (NLMD)", "Denoised Image (RED-CNN)", "Noisy Image with Gaussian Noise"], 
           loc="upper right", fontsize=15, frameon=True)
plt.grid(True)

plt.xlabel("Noise Level", fontsize=15)
plt.ylabel("SSIM", fontsize=15)
custom_xticks = np.arange(0.1, 0.31, 0.02)  
custom_yticks = np.arange(0, 1.4, 0.1)
plt.xticks(custom_xticks, fontsize=15)
plt.yticks(custom_yticks, fontsize=15)


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:\Matthew\Kuliah\Semester 8\Pengolahan Data\Progress code terbaru (4 apr)\Analisis2 - Copy.xlsx"
xls = pd.ExcelFile(file_path)

selected_range = pd.read_excel(xls, sheet_name=0, usecols="M:O", skiprows=0, nrows=4)

x_labels = selected_range.iloc[:, 0].astype(str)
x_labels = [label.replace(" ", "\n") for label in x_labels]
y_values = selected_range.iloc[:, 1].astype(float)
y_errors = selected_range.iloc[:, 2].astype(float)

fig, ax = plt.subplots(figsize=(10, 10), facecolor='white')

hatch_patterns = ['', '//', '\\', 'x']

bars = []
for i in range(len(y_values)):
    bar = ax.bar(x_labels[i], y_values[i], yerr=y_errors[i], capsize=4,
                 edgecolor='black', width=0.5, color='white', hatch=hatch_patterns[i])
    bars.append(bar[0])

plt.ylabel("SSIM", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)


plt.show()
```
